```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm= ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    # temperature=1.0,  # Gemini 3.0+ defaults to 1.0
    # max_tokens=None,
    # timeout=None,
    # max_retries=2,
)
```

```python
res=llm.invoke("Who won in IPL 2024")
print(res.content)
```

```python
#prompt technique
prompts=[
    ("system", "You are an python developer"),
    ("user", "what is a decorator in python explain with example")
]

res=llm.invoke(prompts)
print(res.content)
```
