```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a:int, b:int):
    """
        it will return the sum of two numbers

        Args:
           a:Number One
           b:Number Two
    """
    return a+b
```

```python
@tool
def multiply_numbers(a:int, b:int):
    """
        it will return the multiplication of two numbers

        Args:
           a:Number One
           b:Number Two
    """
    return a*b
```

```python
add_numbers.invoke({"a":12,"b":3})
multiply_numbers.invoke({"a":2, "b":4})
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
agent = create_agent(
    model=llm,
    tools=[add_numbers ,multiply_numbers],
    system_prompt="You are a math teacher you use tools for calculation"
)
```

```python
response=agent.invoke({'messages':[{"role":"user","content":"What is 2+6 ?"}]})
```

```python
print(response)
```

```python
for res in response["messages"]:
    print(res)
    print("\n")
```
