Solution is Structured Output

```python
from pydantic import BaseModel,Field
from typing import List

class ResponseStrecture(BaseModel):
    name:str=Field(description="Complete Name")
    email:str=Field(description="Email Address")
    age:str=Field(description="Age")

structured_llm=llm.with_structured_output(ResponseStrecture)
```

```python
res=structured_llm.invoke(f"Please give me only name and email and age from: {text} ")
```
